```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("darkgrid")

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

## I. The Search for Empirical Truth

```python
df = pd.read_csv('simulated_data.csv')
df.head()
```

```python
# Functions used before
def fit_plot(df1, model):
    features = df1.drop('y', axis=1)
    target = df1.y 
    algo = model.fit(features, target)
    plt.scatter(df1.x, df1.y, color= 'olive', alpha=0.6)
    plt.plot(df1.x, algo.predict(features), 'k--')
    return algo

def fit_polynomial(data, model, power):
    df_copy = data.copy()
    if power == 1:
        algo = fit_plot(df_copy, model)
    else:
        for i in range(2, power+1):
            df_copy['x{}'.format(i)] = np.power(df_copy.x, i)
        algo = fit_plot(df_copy, model)
    return algo
        
```

```python
# New way of defining functions:
# 1) Engineeer polynomials; 2) run the fit and plot function

def engineer_polynomials(data, max_power):
    df_copy= data.copy()
    if max_power == 1:
        return df_copy
    else:
        for i in range(2, max_power + 1):
            df_copy['x{}'.format(i)] = np.power(df_copy.x, i)
        return df_copy

def fit_polynomial(data, model, max_power):
    return fit_plot(engineer_polynomials(data, max_power), model)

    
```

```python
fit_polynomial(df, LinearRegression(), 3)
plt.show()
```

```python
# Mean model
mean_model = np.average(df.y)
print(mean_model)
```

```python
# Linear model
linear_model = fit_polynomial(df, LinearRegression(), 1)
```

```python
# 2nd order polynomial
poly_model_2 = fit_polynomial(df, LinearRegression(), 2)
```

```python
# 3rd order polynomial
poly_model_3= fit_polynomial(df, LinearRegression(), 3)
```

```python
# 5th order polynomial
poly_model_5= fit_polynomial(df, LinearRegression(), 5)
```

```python
# 10th order polynomial
poly_model_10= fit_polynomial(df, LinearRegression(), 10)
```

```python
# Unconstrained tree
tree_uc = fit_plot(df, DecisionTreeRegressor())
```

```python
# Constrained tree --- max depth
tree_maxd3 = fit_plot(df, DecisionTreeRegressor(max_depth= 3))
```

```python
# Constrained tree --- min sample
tree_mins = fit_plot(df, DecisionTreeRegressor(min_samples_leaf= 5))
```

```python
# Unconstrained randome forest
rf_uc = fit_plot(df, RandomForestRegressor(random_state= 123))
```

```python
# Unconstrained randome forest, max depth
rf_maxd3 = fit_plot(df, RandomForestRegressor(random_state= 123, max_depth= 3))
```

```python
# Unconstrained randome forest, min leaf
rf_mins5 = fit_plot(df, RandomForestRegressor(random_state= 123, min_samples_leaf= 5))
```

## II. Unseen Data

```python
# Create test set 
x_test = np.linspace(0, 2*np.pi, 100)

np.random.seed(987)
noise_test = np.random.normal(0,0.5, 100)

y_test = np.sin(x_test) + noise_test
```

```python
test_df = pd.DataFrame({
    'x': x_test,
    'y': y_test
})
test_df.head()
```

```python
# Compare the train sample and test sample
plt.scatter(df.x, df.y, color='olive', alpha=0.6)
plt.scatter(test_df.x, test_df.y, color='maroon')
plt.show
```

```python
features_test = test_df.drop('y', axis= 1)
target_test = test_df.y
```

## III. Performance Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

yactual = [0, 1, 2, 3, 4]
yhat = [2, 2, 2, 2, 2]
```

```python
# MSE
mean_squared_error(yhat, yactual)
```

```python
# MAE
mean_absolute_error(yhat, yactual)
```

## IV. Model Evaluation & Selection

```python
# Store predictions

# Meanmodel
yhat_mean = [mean_model] * len(features_test)

# Linear model
yhat_linear = linear_model.predict(features_test)

# Linear model power 2
yhat_linear2 = poly_model_2.predict(engineer_polynomials(features_test, 2))

# Linear model power 3
yhat_linear3 = poly_model_3.predict(engineer_polynomials(features_test, 3))

# Linear model power 5
yhat_linear5 = poly_model_5.predict(engineer_polynomials(features_test, 5))

# Linear model power 10
yhat_linear10 = poly_model_10.predict(engineer_polynomials(features_test, 10))

# tree, unconstrained
yhat_tree_uc = tree_uc.predict(features_test)

# tree, max depth 3
yhat_tree_maxd3 = tree_maxd3.predict(features_test)

# tree, min leaf 5
yhat_tree_mins5 = tree_mins.predict(features_test)

# random forest, unconstrained
yhat_rf_uc = rf_uc.predict(features_test)

# randome forest, max depth
yhat_rf_maxd3 = rf_maxd3.predict(features_test)

# random forest, min leaf
yhat_rf_mins5 = rf_mins5.predict(features_test)
```

```python
# Print metrics 
print("Mean Sqaured Error:")
yhats= [yhat_mean, yhat_linear, yhat_linear2, yhat_linear3, yhat_linear5, yhat_linear10,
    yhat_tree_uc, yhat_tree_maxd3, yhat_tree_mins5, yhat_rf_uc, yhat_rf_maxd3, yhat_rf_mins5] 
models= ['mean model', 'linear model', 'poly2', 'ploy3', 'ploy5', 'ploy10',
    'unconstrained tree', 'tree, max depth 3', 'tree, min leaf 5', 
    'randome forest, unconstrained', 'random forest, max depth 3', 'random forest, min leaf 5']
for (i,j) in zip(yhats, models):
    print(np.round(mean_squared_error(i, target_test), 3), j)
```

1. The MSE increases from 3 to 10, so we are overfitting after 3 by adding model complexity 
2. The unconstrained tree performs really bad by overfitting. The randome forest performs better by smoothing out predictions

```python
# Try using MAE instead 
for (i,j) in zip(yhats, models):
    print(np.round(mean_absolute_error(i, target_test), 3), j)
```

Aha! As it turns out, the minimum, 25th-percentile, median, and 75th-percentile absolute errors are comparable (only slightly higher) than those for the decision tree model. However, the **maximum** error is much, much higher than that of the decision tree model.

Now that we see this information, if we turn around and think a bit more about the problem, it actually begins to make a lot of sense. If you remember from trigonometry, the tangent function is **asymptotic**. As $x$ approaches certain values at regular intervals, then $y$ actually approaches $\infty$ or $-\infty$.

Practically speaking, this means that we will end up with extremely large or extremely small values of $y$ that **appear as outliers** to many machine learning algorithms. Thus, a random forest, which "smoothes out" the predictions, could end up severely punished by predictions near those asymptotes.

Let's see what we mean:
